# Load NRCan Data from Relevant Region

```python
import pandas as pd
from datetime import datetime
import os
```

```python
load_list = []
for item in os.listdir("data"):
    if item[:15] == 'daily_20220922T':
        load_list.append(item)
load_list
```

```python
dfs = []
for item in load_list:
    df = pd.read_csv(f"./data/{item}", low_memory=False, header=1)
    dfs.append(df)
dfs[-1]
```

```python
df_all = pd.concat(dfs)
df_all.reset_index(drop=True, inplace=True)
df_all
```

# Check for Data to Fill Null Discharge Values

```python
percent_missing = combined_df["discharge"].isnull().sum() * 100 / len(combined_df)
percent_missing
```

```python
month_val_dict = {1:'Jan', 2:'Feb', 3:'Mar', 4: 'Apr', 5:'May', 6: 'Jun', 
                  7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}

df_len = len(combined_df.index)

filled_count = 0

counter = 0

def check_for_new_discharge(x, df_all):
    global counter
    global filled_count
    discharge = x["discharge"]
    if pd.isnull(discharge):
        official_id = x["Official_ID"]
        year = int(x["time"][:4])
        month = month_val_dict[int(x["time"][5:7])]
        day = int(x["time"][8:10])
        new_discharge_df = df_all[(df_all[" ID"]==official_id) & (df_all["YEAR"]==year) & (df_all["DD"]==day) & (df_all["PARAM"]==1)][month]
        if not new_discharge_df.empty:
            new_discharge = new_discharge_df.iloc[0]
            if new_discharge:
                filled_count += 1
            discharge = new_discharge   
    if counter % 100000 == 0:
        now = datetime.now()
        dt_string = now.strftime("%Y/%m/%d %H:%M:%S")
        print(f"Completed {counter} lines of {df_len} at {dt_string}.")
    counter += 1    
    return discharge

combined_df["discharge"] = combined_df.apply(lambda x: check_for_new_discharge(x, df_all), axis = 1)

print(filled_count)
combined_df
```

```python
percent_missing = combined_df["discharge"].isnull().sum() * 100 / len(combined_df)
percent_missing
```
