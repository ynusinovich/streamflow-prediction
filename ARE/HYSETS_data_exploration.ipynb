```python
# load watershed properties
wsp = pd.read_csv('/content/drive/My Drive/Streamflow Data/HYSETS_watershed_properties.txt')
# useless drop columns or columns with lots of NaNs
wsp.drop(columns=['Drainage_Area_GSIM_km2', 'Flag_GSIM_boundaries', 'Flag_Artificial_Boundaries'], inplace=True)

# check for missing values
print('Missing data (before trimming):')
str_cols = []
for col in wsp.columns:
    try: 
        vals = wsp[col].values
        print(col, np.isnan(vals).mean(), np.nanmin(vals), np.nanmax(vals), np.nanmean(vals), np.nanstd(vals), )
    except TypeError:
        str_cols.append(col)
print('String Columns:', *str_cols)

# drop rows (watersheds) with missing meta data
orig_len = wsp.shape[0]
wsp.dropna(axis='index', inplace=True)
wsp_len = wsp.shape[0]

# check how much was dropped
print('\nRemaining data (after trimming):',wsp_len/orig_len)
```
